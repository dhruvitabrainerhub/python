```python
#data filtering
data = [
    ["Riya", 22, 40000],
    ["Aman", 17, 20000],
    ["Neha", 25, 50000]
]
def eligible(data):
    valid=[]
    for d in data:
        if d[1]>=21 and d[2]>=35000:
            valid.append(d[0])
        return valid
result = print(eligible(data))
```

```python
#zip method 
names ="A","B"
salary = 20000,40000
for n , s in zip(names,salary):
    print(n,s)
```

```python
#enumerate use for indexing list
lst =["apple","banana","mango"]
for index,name in enumerate(lst):
    print(index,name)
```

```python
# reverse string
s = "python"
rev = ""
for i in s:
    rev = i+rev
print(rev)
```

```python
# vowels counting
s = "education"
vowels ="aeiou"
count = 0
for ch in s:
    if ch in vowels:
        count+=1
print(count)
```

```python
#find a maximum number
lst = [10, 20, 30, 40]
max_num = lst[0]
for i in lst:
    if i > max_num:
        max_num = i
print(max_num)
```

```python
#square for list
lst = [2, 4, 6, 8]
square =(l*l for l in lst )
for s in square:
    print(s)
```

```python
#pattern print
for i in range(1,6):
    for j in range(i):
        print("*",end="")
    print()
```
